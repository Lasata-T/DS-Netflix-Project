```python
numbered_regions = gdp_data2.insert(5, 'Numbered Regions', 0) # zero is placeholder
gdp_data2['Numbered Regions'] # do NOT just insert and put placeholder 
# try to add correct values right as you make the new column
```

```python
for i in gdp_data2['Region']:
    if i == ('South Asia'):
        gdp_data2['Numbered Regions'].replace(0, 1)
    
    if i == ('Sub-Saharan Africa'):
        gdp_data2['Numbered Regions'].replace(0, 2)
    
    if i == ('Europe & Central Asia'):
        gdp_data2['Numbered Regions'].replace(0, 3)
    
    if i == ('Middle East & North Africa'):
        gdp_data2['Numbered Regions'].replace(0, 4)
        
    if i == ('Latin America & Caribbean'):
        gdp_data2['Numbered Regions'].replace(0, 5)
    
    if i == ('East Asia & Pacific'):
        gdp_data2['Numbered Regions'].replace(0, 6)
    
    if i == ('North America'):
        gdp_data2['Numbered Regions'].replace(0, 7)
```
